```python
import seaborn as sns
p = ["#A04747" , "#D8A25E","#EEDF7A"]
c = sns.countplot(data=df , x='AgeRange', palette=p,hue='Gender')
for patch in c.patches:
    patch.set_edgecolor('black') 
    patch.set_linewidth(2)

plt.title("Distribution of Age Ranges by Gender")
plt.xlabel("Age Range", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt 

# Define colors
c = ["#243642", "#629584"]

# Count values in the 'Attrition' column
Attrition_counts = df['Attrition'].value_counts()

# Create the pie chart
plt.pie(
    Attrition_counts,
    labels=Attrition_counts.index,  # Use index for proper labels
    shadow=False,
    labeldistance=1.2,
    explode=(0.1, 0),  # Use tuple for explode
    colors=c,
    autopct='%.1f%%',
    wedgeprops=dict(edgecolor='black'),
    textprops={'fontsize': 15,'color': '#CB6040'}  # Font size for text
)

# Set the title
plt.title('Distribution of Attrition', fontsize=18)  # Set title font size
plt.show()
```

```python
c = ["#243642" , "#629584"]

overtime_distance = df.groupby('OverTime')['DistanceFromHome'].mean()

plt.pie(overtime_distance, labels=overtime_distance.index, autopct='%1.1f%%',shadow=True,colors=c,textprops={'color': 'white'}, 
        wedgeprops={'edgecolor': 'black', 'linewidth': 2})
plt.title('Average Distance from Home for Overtime Status')

plt.show()
```

```python
c = ["#243642" , "#629584"]


import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))  # Set the figure size

# Create the barplot
barplot = sns.barplot(data=df, x='JobSatisfaction', y='DistanceFromHome', hue='Attrition', estimator='mean', errorbar=None,palette=c)
for patch in barplot.patches:
    patch.set_edgecolor('white') 
    patch.set_linewidth(4)
# Set custom font sizes for axes labels and ticks
plt.title('Job Satisfaction vs. Avg Distance From Home by Attrition', fontsize=20)
plt.xlabel('Job Satisfaction', fontsize=16)
plt.ylabel('Distance From Home (Miles)', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Display the plot
plt.show()
```
